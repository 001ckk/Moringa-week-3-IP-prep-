```python
#key to the geo data

key1 = pd.read_excel('cells_geo_description.xlsx')
key1
```

```python
# Loading the geo data
# Making the geo columns small letters (to be able to combine with the Telcoms data)

geo = pd.read_csv("cells_geo.csv",delimiter = ';')
new_column = ['unnamed','villes','status','localisation','decoupzone','zonename','longitude','latitude','region','area', 'cell_id','site_code'] 
geo.columns = new_column
geo
```

```python
#Loading the Telcom_datasets
```

```python
#Description of the Telcoms_dataset key

key2 = pd.read_excel('CDR_description.xlsx')
key2
```

```python
# Importing Telcom data sets from excel in csv form
# Data_set1

table1 = pd.read_csv('Telcom_dataset1.csv')
table2 = pd.read_csv('Telcom_dataset2.csv')
table3 = pd.read_csv('Telcom_dataset2.csv')
```

```python
# Changing column names to similar names in Data_set 1,2 amd 3

new_columns = ['product', 'value', 'date_time', 'cell_on_site', 'dw_a_number_int',
'dw_b_number_int', 'country_a', 'country_b', 'cell_id', 'site_id']

table1.columns = new_columns
table2.columns = new_columns
table3.columns = new_columns
```

```python
# Merging the three data sets

Telcom_merged = pd.concat([table1, table2, table3], axis= False)
Telcom_merged.head(2)
```

```python
# Changing date_time format.


Telcom_merged['date_time'] = pd.to_datetime(Telcom_merged['date_time'])
Telcom_merged
```

```python
# dropping country_a and country_b columns 

new = Telcom_merged.drop(['country_a', 'country_b'], axis = True)
new
```

```python
#Removing the duplicate datas 
```

```python
# Removing the duplicate datas in the Telcom data 

Telcom_merged1 = new.drop_duplicates()
Telcom_merged1.head(2)
```

```python
# Removing duplicates in the CDR data

geo1 = geo.drop_duplicates()
geo1.head(2)
```

```python
# Replacing null values with 0's
# geo data
geo1.fillna(0)
```

```python
# Telcom_merged1

Telcom_merged1.fillna(0)
```

```python
# merging the CDR data and Telcom Data and selecting only the important columns

Tel_geo = pd.merge(Telcom_merged1,geo1, how='inner', on=['cell_id'])

Tel_geo[['product','value','date_time','cell_id','villes']].info()
```
